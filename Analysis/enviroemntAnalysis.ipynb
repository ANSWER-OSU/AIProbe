```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file paths
file_names = [
    "/Users/rahil/Documents/GitHub/AIProbe/csharp/Result/re/6161/_values.csv",
    "/Users/rahil/Documents/GitHub/AIProbe/csharp/Result/re/6161/random_values.csv"  # Add more file paths here
]

# Define a list of contrasting colors
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'cyan']

# Initialize the plot
plt.figure(figsize=(10, 6))

# To store combined data for overlap detection
all_data = []

# Iterate over each file and plot its data with contrasting colors
for idx, file_name in enumerate(file_names):
    data = pd.read_csv(file_name)
    
    # Append data to the combined list for overlap detection
    all_data.append(data)
    
    # Plot data points
    plt.scatter(data['Grid'], data['Lava'], color=colors[idx % len(colors)], alpha=0.8, label=f'Data from {file_name.split('/')[-1]}')

# Combine all data into a single DataFrame
combined_data = pd.concat(all_data)

# Detect duplicate points (overlapping points)
duplicates = combined_data.duplicated(subset=['Grid', 'Lava'], keep=False)

# Highlight overlapping points with circles
for _, row in combined_data[duplicates].iterrows():
    plt.gca().add_artist(plt.Circle((row['Grid'], row['Lava']), 0.1, edgecolor='black', facecolor='none', linewidth=1.5))

# Customize the graph
plt.title('Grid vs. Lava (Overlapping Points Highlighted)', fontsize=14)
plt.xlabel('Grid', fontsize=12)
plt.ylabel('Lava', fontsize=12)
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(range(int(combined_data['Grid'].min()), int(combined_data['Grid'].max() + 1)))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Function for Latin Hypercube Sampling
def lhs_sampling(samples, dimensions):
    """
    Generate Latin Hypercube Samples for given dimensions and number of samples.
    """
    result = np.zeros((samples, dimensions))
    for i in range(dimensions):
        points = np.random.permutation(np.linspace(0, 1, samples, endpoint=False) + np.random.uniform(0, 1 / samples, samples))
        result[:, i] = points
    return result

# Define the ranges for grid and lava
num_samples = 10  # Number of samples
grid_min, grid_max = 1, 5  # Range for grid

# Generate LHS samples for grid (integer values)
lhs_samples = lhs_sampling(num_samples, 1)
grid_values = np.round(grid_min + (grid_max - grid_min) * lhs_samples[:, 0]).astype(int)

# Generate corresponding lava values (integer values)
lava_values = [np.random.randint(0, grid**2 + 1) for grid in grid_values]

# Create a DataFrame to display the sampled values
data = pd.DataFrame({
    'Grid': grid_values,
    'Lava': lava_values
})

# Display the DataFrame
print(data)

# Optionally save to a CSV file
data.to_csv('grid_lava_samples_int.csv', index=False)


import matplotlib.pyplot as plt

# Ensure both 'Grid' and 'Lava' are integers
data['Grid'] = data['Grid'].astype(int)
data['Lava'] = data['Lava'].astype(int)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(data['Grid'], data['Lava'], s=100, alpha=0.7, edgecolor='k', marker='o')

# Add labels and title
plt.title('Grid vs Lava Values', fontsize=14)
plt.xlabel('Grid', fontsize=12)
plt.ylabel('Lava', fontsize=12)

# Customize the grid and ticks
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(data['Grid'].min(), data['Grid'].max() + 1), fontsize=10)
plt.yticks(range(data['Lava'].min(), data['Lava'].max() + 1), fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
