```python
import numpy as np
import pandas as pd

# Function for Latin Hypercube Sampling
def lhs_sampling(samples, dimensions):
    """
    Generate Latin Hypercube Samples for given dimensions and number of samples.
    """
    result = np.zeros((samples, dimensions))
    for i in range(dimensions):
        points = np.random.permutation(np.linspace(0, 1, samples, endpoint=False) + np.random.uniform(0, 1 / samples, samples))
        result[:, i] = points
    return result

# Define the ranges for grid and lava
num_samples = 10  # Number of samples
grid_min, grid_max = 1, 5  # Range for grid

# Generate LHS samples for grid (integer values)
lhs_samples = lhs_sampling(num_samples, 1)
grid_values = np.round(grid_min + (grid_max - grid_min) * lhs_samples[:, 0]).astype(int)

# Generate corresponding lava values (integer values)
lava_values = [np.random.randint(0, grid**2 + 1) for grid in grid_values]

# Create a DataFrame to display the sampled values
data = pd.DataFrame({
    'Grid': grid_values,
    'Lava': lava_values
})

# Display the DataFrame
print(data)

# Optionally save to a CSV file
data.to_csv('grid_lava_samples_int.csv', index=False)


import matplotlib.pyplot as plt

# Ensure both 'Grid' and 'Lava' are integers
data['Grid'] = data['Grid'].astype(int)
data['Lava'] = data['Lava'].astype(int)

# Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(data['Grid'], data['Lava'], s=100, alpha=0.7, edgecolor='k', marker='o')

# Add labels and title
plt.title('Grid vs Lava Values', fontsize=14)
plt.xlabel('Grid', fontsize=12)
plt.ylabel('Lava', fontsize=12)

# Customize the grid and ticks
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(data['Grid'].min(), data['Grid'].max() + 1), fontsize=10)
plt.yticks(range(data['Lava'].min(), data['Lava'].max() + 1), fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import ks_2samp, levene, f
from statsmodels.multivariate.manova import MANOVA

# Step 2: Load the datasets
random_samples = pd.read_csv('/Users/rahil/Documents/GitHub/AIProbe/csharp/Result/re/6161/random_values.csv')
Lhs = pd.read_csv('/Users/rahil/Documents/GitHub/AIProbe/csharp/Result/re/6161/improvedLHS_values.csv')

# Step 3: Add group labels for combined analysis
random_samples['Group'] = 'Random'
Lhs['Group'] = 'Lhs'

# Combine datasets for MANOVA
combined_data = pd.concat([random_samples, Lhs])

# Step 4: Hotellings T-Squared Test (Alternative Method)
def hotelling_t2_test(data1, data2):
    n1, p = data1.shape
    n2 = data2.shape[0]
    mean_diff = data1.mean(axis=0) - data2.mean(axis=0)
    pooled_cov = ((n1 - 1) * np.cov(data1, rowvar=False) + (n2 - 1) * np.cov(data2, rowvar=False)) / (n1 + n2 - 2)
    t2_stat = n1 * n2 / (n1 + n2) * mean_diff.T @ np.linalg.inv(pooled_cov) @ mean_diff
    f_stat = t2_stat * (n1 + n2 - p - 1) / (p * (n1 + n2 - 2))
    p_value = 1 - f.cdf(f_stat, p, n1 + n2 - p - 1)
    return {"T2 Statistic": t2_stat, "F-Statistic": f_stat, "P-value": p_value}

data1 = random_samples[['Grid', 'Lava']].values
data2 = Lhs[['Grid', 'Lava']].values
hotelling_result = hotelling_t2_test(data1, data2)

# Step 5: MANOVA
manova = MANOVA.from_formula('Grid + Lava ~ Group', data=combined_data)
manova_result = manova.mv_test()

# Step 6: Kolmogorov-Smirnov Test
data1_flattened = data1.flatten()
data2_flattened = data2.flatten()
ks_stat, ks_p_value = ks_2samp(data1_flattened, data2_flattened)

# Step 7: Levenes Test for Homogeneity of Variances
levene_stat_x, levene_p_x = levene(random_samples['Grid'], Lhs['Grid'])
levene_stat_y, levene_p_y = levene(random_samples['Lava'], Lhs['Lava'])

# Output results
results = {
    "Hotelling's T-Squared Test": hotelling_result,
    "MANOVA Result": str(manova_result),  # Convert MANOVA output to string for display
    "Kolmogorov-Smirnov Test": {"KS Statistic": ks_stat, "P-value": ks_p_value},
    "Levene's Test": {
        "Grid": {"Statistic": levene_stat_x, "P-value": levene_p_x},
        "Lava": {"Statistic": levene_stat_y, "P-value": levene_p_y}
    }
}

# Display results
for test, result in results.items():
    print(f"{test}:\n{result}\n")
```

```python
# Function to compute Mahalanobis distances for all points in a dataset
def compute_mahalanobis(data):
    """
    Compute the Mahalanobis distance for each point in the dataset.
    
    Parameters:
        data (numpy.ndarray): Standardized dataset (rows are samples, columns are features).
    
    Returns:
        numpy.ndarray: Mahalanobis distances for each point.
    """
    mean_vector = np.mean(data, axis=0)  # Mean vector
    covariance_matrix = np.cov(data, rowvar=False)  # Covariance matrix
    inv_covariance_matrix = np.linalg.inv(covariance_matrix)  # Inverse covariance matrix
    
    distances = []
    for i in range(data.shape[0]):
        diff = data[i] - mean_vector  # Difference from the mean
        md = np.sqrt(diff.T @ inv_covariance_matrix @ diff)  # Mahalanobis distance formula
        distances.append(md)
    
    return np.array(distances)

# Compute Mahalanobis distances for both datasets
file1_mahalanobis = compute_mahalanobis(file1_standardized)
file2_mahalanobis = compute_mahalanobis(file2_standardized)

# Print results
print(f"File 1 Mahalanobis Distances: {file1_mahalanobis}")
print(f"File 2 Mahalanobis Distances: {file2_mahalanobis}")
```
