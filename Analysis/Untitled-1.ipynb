```python
import csv
import re

def parse_csv_with_labels(file_path):
    results = []

    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        header = next(csv_reader)  # Skip the header row

        for row_number, row in enumerate(csv_reader, start=1):
            try:
                # Extract columns
                environment_name = row[0]
                environment_type = row[1]
                attributes = row[2]
                agents = row[3]
                objects = row[4]

                # Parse each section and structure the data with labels
                attribute_data = parse_attributes(attributes)
                agent_data = parse_agents(agents, row_number)
                object_data = parse_objects(objects, row_number)

                # Combine all labeled data into a single structured result
                result = {
                    "EnvironmentName": environment_name,
                    "EnvironmentType": environment_type,
                    **attribute_data,
                    **agent_data,
                    **object_data
                }
                results.append(result)

            except Exception as e:
                print(f"Error processing row {row_number}: {row}")
                print(f"Exception: {e}")

    return results

def parse_attributes(data):
    """
    Extract labeled attributes from the attribute string.
    Example: "Grid=7 (Type=int); Lava=8 (Type=int)" -> {"Grid": 7, "Lava": 8}
    """
    try:
        attributes = re.findall(r'(\w+)=([\d.]+)', data)
        return {attr: int(value) for attr, value in attributes}
    except Exception as e:
        print(f"Error parsing attributes: {data}")
        print(f"Exception: {e}")
        return {}

def parse_agents(data, row_number):
    """
    Extract labeled agent attributes from the agent string.
    Example: "Agent_1: [X=4 (Type=int), Y=3 (Type=int), Angle=3 (Type=int)]"
    """
    agents = {}
    try:
        agent_matches = re.findall(r'Agent_\d+: \[(.*?)\]', data)
        for i, match in enumerate(agent_matches, 1):
            for key, value in re.findall(r'([XYAngle])=([\d.]+)', match):
                agents[f"Agent_{row_number}_{i}_{key}"] = int(value)  # Row-specific unique keys
    except Exception as e:
        print(f"Error parsing agents in row {row_number}: {data}")
        print(f"Exception: {e}")
    return agents

def parse_objects(data, row_number):
    """
    Extract labeled object attributes from the object string.
    Example: "Object_9 (Type=Lava): [X=2 (Type=int), Y=4 (Type=int)]"
    """
    objects = {}
    try:
        object_matches = re.findall(r'Object_\d+ \(Type=\w+\): \[(.*?)\]', data)
        for i, match in enumerate(object_matches, 1):
            for key, value in re.findall(r'([XY])=([\d.]+)', match):
                objects[f"Object_{row_number}_{i}_{key}"] = int(value)  # Row-specific unique keys
    except Exception as e:
        print(f"Error parsing objects in row {row_number}: {data}")
        print(f"Exception: {e}")
    return objects

def save_to_csv(data, output_file_path):
    try:
        with open(output_file_path, mode='w', newline='') as file:
            # Gather all unique keys across all rows
            fieldnames = set()
            for row in data:
                fieldnames.update(row.keys())
            fieldnames = sorted(fieldnames)

            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(data)
    except Exception as e:
        print(f"Error saving to CSV: {output_file_path}")
        print(f"Exception: {e}")

# Example Usage
file_path = "/Users/rahil/Documents/GitHub/AIProbe/csharp/results/Result_LavaEnv_6161/sampled_values.csv"
parsed_data = parse_csv_with_labels(file_path)

# Print structured results for debugging
for idx, row in enumerate(parsed_data, start=1):
    print(f"Row {idx}: {row}")

# Save the parsed data
output_path = "/Users/rahil/Documents/GitHub/AIProbe/csharp/results/Result_LavaEnv_6161/output_vectors.csv"
save_to_csv(parsed_data, output_path)
```

```python
import pandas as pd

# Specify the path to the CSV file
csv_file_path = '/Users/rahil/Documents/GitHub/AIProbe/csharp/results/Result_LavaEnv_6161/sampled_values.csv'

# Create a function to clean trailing commas and pad rows
def clean_and_pad_csv(file_path, output_path):
    rows = []
    max_columns = 0

    # Read the file line by line
    with open(file_path, 'r') as file:
        for line in file:
            # Strip whitespace and replace trailing commas with '0'
            clean_line = line.strip().rstrip(',')
            clean_line = clean_line.replace(',,', ',0,')  # Replace double commas with a single 0
            if clean_line.endswith(','):
                clean_line += '0'  # Append a zero if it still ends with a comma
            values = clean_line.split(',')
            rows.append(values)
            # Update the maximum column count
            max_columns = max(max_columns, len(values))

    # Pad rows to ensure all rows have the same number of columns
    padded_rows = [row + ['0'] * (max_columns - len(row)) for row in rows]

    # Convert to a DataFrame
    df_padded = pd.DataFrame(padded_rows)

    # Save the cleaned and padded DataFrame to a new file
    df_padded.to_csv(output_path, index=False, header=False)
    print(f"Cleaned and padded CSV file saved to {output_path}")

# Path to save the output
output_csv_path = '/Users/rahil/Documents/GitHub/AIProbe/csharp/results/Result_LavaEnv_6161/output_vectors.csv'

# Run the function
clean_and_pad_csv(csv_file_path, output_csv_path)
```
